```
import numpy as np
import matplotlib.pyplot as plt
from math import comb

# Definición de parámetros para la simulación
clientes_diarios = 20  # Total de clientes por jornada
probabilidad_disponibilidad = 0.70  # Probabilidad de encontrar el producto
cantidad_dias = 7  # Días de prueba

import numpy as np
import matplotlib.pyplot as plt

# Definición del parámetro para la distribución exponencial (tasa de carga por vehículo)
tasa_carga = 1 / 15  # Promedio de carga de 15 minutos

# Función para simular un tiempo de carga
def simular_tiempo_carga(tasa):
    numero_aleatorio = np.random.uniform(0, 1)  # Generar un número aleatorio
    # Aplicar el método de transformación inversa
    tiempo = -np.log(1 - numero_aleatorio) / tasa
    return tiempo

# Simular tiempos de carga durante 12 horas (48 ciclos)
tiempo_por_carga = [simular_tiempo_carga(tasa_carga) for _ in range(48)]

# Mostrar los tiempos generados
print("Tiempos de carga registrados:", tiempo_por_carga)

# Visualización de la distribución de tiempos de carga
plt.hist(tiempo_por_carga, bins=10, edgecolor='black', alpha=0.7)
plt.title('Simulación de la Distribución de Tiempos de Carga (Exponencial)')
plt.xlabel('Duración de la carga (minutos)')
plt.ylabel('Frecuencia')
plt.show()

```
